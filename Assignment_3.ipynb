# FizzBuzz
Write a program that prints the numbers from 1 to 100. But for multiples of three print “Fizz” instead of the number and for the multiples of five print “Buzz”. For numbers that are multiples of both three and five print “FizzBuzz”.

````python
for i in range(100):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
````

# Tic-Tac-Toe Game Winner
Given a board of a tic-tac-toe game, find the winner. A sample board is given as follows:
```py
board = [['O', 'X', 'X'],
         ['X', 'X', 'O'],
         ['O', 'X', 'X']]
```


````python
board = [['O', 'X', 'X'],
         ['X', 'X', 'O'],
         ['O', 'X', 'X']]

n = len(board)

lines = []

# rows and columns
for i in range(n):
    lines.append(board[i])                       # row i
    lines.append([board[r][i] for r in range(n)])# col i

# diagonals
lines.append([board[i][i] for i in range(n)])
lines.append([board[i][n-1-i] for i in range(n)])

for line in lines:
    s = set(line)
    if len(s) == 1 and line[0] in {"X", "O"}:
        print(line[0])
        break;
else:
    print("draw, no winner")
    
        
        
 # "X" or "O"
````

# Password Strenghth Meter
Write a Python program that reads passwords from users. For the given password, tell users whether the password is **weak**, **okay**, or **strong**. Here are the scoring rules for a password:

- Length
  - +2 if length ≥ 12
  - +1 if length is 8–11
  - +0 otherwise
- Character variety
  - +1 if it has both lowercase [a-z] and uppercase [A-Z]
  - +1 if it has at least one digit [0-9]
  - +1 if it has at least one special from this set only: !@#$%^&*
- Penalties
  - −2 if it contains any whitespace (space, tab, etc.)
  - −3 if it contains three or more identical characters in a row (e.g., aaa, 1111, ***)
- Final Result
  - weak if score < 3
  - okay if 3 ≤ score ≤ 4
  - strong if score ≥ 5

Test your program with the following passwords:
- P@ssw0rd
- hunter2
- CorrectHorseBatteryStaple
- abcde Fgh
- A!!!A!!!A!!!

**Hint**: You may find `str.islower()`, `str.isupper()`, `str.isdigit()`, and `str.isspace()` useful.

````python
SPECIALS = set("!@#$%^&*")
def has_run_ge3(s: str) -> bool:
    run = 1
    for i in range(1, len(s)):
        if s[i] == s[i-1]:
            run += 1
            if run >= 3:
                return True
        else:
            run = 1
    return False

def password_score(pw: str) -> int:
    score = 0

    # --- Length ---
    n = len(pw)
    if n >= 12:
        score += 2
    elif 8 <= n <= 11:
        score += 1
    # else +0

    # --- Character variety ---
    has_lower = any(c.islower() for c in pw)
    has_upper = any(c.isupper() for c in pw)
    has_digit = any(c.isdigit() for c in pw)
    has_special = any(c in SPECIALS for c in pw)

    if has_lower and has_upper:
        score += 1
    if has_digit:
        score += 1
    if has_special:
        score += 1

    # --- Penalties ---
    # - whitespace
    if any(c.isspace() for c in pw):
        score -= 2

    # - three or more identical in a row
    #   e.g. 'aaa', '1111', '***'
    if has_run_ge3(pw):
        score -= 3

    return score


def password_label(score: int) -> str:
    if score < 3:
        return "weak"
    if 3 <= score <= 4:
        return "okay"
    return "strong"


def assess(pw: str) -> str:
    s = password_score(pw)
    return f"{pw!r} → score={s}, {password_label(s)}"


# ---- Demo with the given test passwords ----
tests = [
    "P@ssw0rd",
    "hunter2",
    "CorrectHorseBatteryStaple",
    "abcde Fgh",
    "A!!!A!!!A!!!",
]

for t in tests:
    print(assess(t))
````

# Vacuum Robot

You get an $n\times n$ map made of:
- \# = wall (can’t go through)
- . = empty floor (can move)
- S = where the robot starts (treated like empty floor)
You’re told which way the robot is facing (N, E, S, or W).

You get a string of commands:
- L turn left 90°
- R turn right 90°
- F try to move one step forward
If a forward step would hit a wall or leave the map, the robot remains where it is. Its health condition decreases by 1.
After all commands, print: row col orientation (rows/cols are 1-based, top-left is (1,1)), as well as robot's health condition.
**Note**: The minimum health condition is 0. Once the robot's health reaches 0, it cannot accept commands any more. 

Test your program with the following grids and commands:

- Grid 1
    ```py
    #####
    #...#
    #.#S#
    #...#
    #####
    ```
    Commands: FFLFFRFFF; assume the robot initially facing east (E) with initial health 3.

- Grid 2
    ```py
    #######
    #..#..#
    #.#.#S#
    #.#...#
    #..####
    #.....#
    #######
    ```
    Commands: FLLFFFRFFLFFRFF; assume the robot initially facing west (W) with initial health 10.

````python
def vacuum_robot(grid, commands, start_dir, health):
    # 方向顺序（顺时针）
    dirs = ["N", "E", "S", "W"]
    # 每个方向对应的移动偏移
    move = {"N": (-1, 0), "E": (0, 1), "S": (1, 0), "W": (0, -1)}
    
    n = len(grid)
    
    # 找起点 S
    for r in range(n):
        for c in range(n):
            if grid[r][c] == "S":
                row, col = r, c

    dir_idx = dirs.index(start_dir)

    for cmd in commands:
        if health == 0:
            break

        if cmd == "L":
            dir_idx = (dir_idx - 1) % 4
        elif cmd == "R":
            dir_idx = (dir_idx + 1) % 4
        elif cmd == "F":
            dr, dc = move[dirs[dir_idx]]
            new_r, new_c = row + dr, col + dc

            # 扣血
            health -= 1

            # 判断是否能走
            if 0 <= new_r < n and 0 <= new_c < n and grid[new_r][new_c] != "#":
                row, col = new_r, new_c

    # 输出结果（行列从1开始）
    print(row + 1, col + 1, dirs[dir_idx], health)

grid1 = [
    "#####",
    "#...#",
    "#.#S#",
    "#...#",
    "#####"
]
vacuum_robot(grid1, "FFLFRFFFF", "E", 3)

grid2 = [
    "#######",
    "#...#.#",
    "#.#..S#",
    "#.#...#",
    "#..####",
    "#......#",
    "#######"
]
vacuum_robot(grid2, "FLLFFRFRFLFFRFF", "W", 10)
````
